```python
# Import pandas with alias
import pandas as pd
```

```python
# to see the data types of the columns:
print(movies.dtypes)
```

```python
"""
If a variable appears with the wrong data type, we can change it with the .astype() function.
The .astype() function can be used to convert between a numerical data types, including:
int32 int64, float32 float64, object, string, bool.

Note : However, some data types require all values to be filled in. For example, you cannot convert between a float and an int if there are any null values.
"""
movies["type"] = movies["type"].astype("string")
print(movies.dtypes)
```

```python
"""
1. Run the code in the workspace to view the first five rows of the movies dataframe.
2. Check the data types of all the numerical variables in the movies DataFrame with .dtypes. Make sure to print the data types to the console.
Are there any mismatches between data types and variable types?
"""
print(movies.head())
print(movies.dtypes)  # Since dtype is an attribute, we used it without any parentheses.
```

```python
# This will cause an error.If we didn’t check for null values in advance, this error message might be the first time we find out that we have null values.
movies["cast_count"] = movies["cast_count"].astype("int64")
```

```python
"""
Turns out we have to fill in the missing values. We actually don’t have a good way to know how many people are in each movie, so we will fill in the missing values with 0. We know that no movie has zero cast members, so this will logically be very odd, but will allow us to preserve the integrity of the data while converting the values.
"""
movies["cast_count"].fillna(0, inplace=True)
print(movies.head()) # check for Nan values.
```

```python
"""
Note 1: Nominal variables are often represented by the object data type. Columns in the object data type can contain any combination of values, including strings, integers, booleans, etc. This means that string operations like .lower() are not possible on object columns.

Note 2: Nominal variables are also represented by the string data type. However, Pandas usually guesses object rather than string, so if you want a column to be a string, you will likely have to explicitly tell pandas to make it a string. This is most important if you want to do string manipulations on a column like .lower().

Note 3: Ordinal variables should be represented as objects, but pandas often guesses int since they are often encoded as whole numbers.

Note 4: Binary variables can be represented as bool, but pandas often guesses int or object data types."""
```

```python
"""
We have a lot to change in our cereal dataset, so let’s go through them one by one.
"""
print(movies.head())
print(movies.dtypes)
movies["title"] = movies["title"].astype("string")
movies["country"] = movies["country"].astype("string")
movies["cast_count"] = movies["cast_count"].astype("int64")
```

```python
"""
print the first five rows of the movies DataFrame along with the unique values of the rating column.
"""
print(movies.head())
print(movies["rating"].unique())
```

```python
"""
Change the data type of the rating variable to categorical. Set an order of [‘NR’, ‘G’ , ‘PG’, ‘PG-13’, ‘R’].
"""
movies["rating"] = pd.Categorical(movies["rating"], ["NR", "G", "PG", "PG-13", "R"], ordered=True)
print(movies["rating"].unique())
```

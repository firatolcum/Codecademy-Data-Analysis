### Plotting Bar Charts Using Seaborn
Letâ€™s practice plotting our own bar chart. In plot.py, there is a csv file loaded called games.csv. This dataset contains information about various chess games, including:
- Number of turns the game took
- Ending result (checkmate/draw/resign)
- Who won (player with white pieces or black pieces)
- Other variables

Plot the counts of each value in `victory_status` using the `.countplot()` method. Be sure to show your plot after using the method.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("games.csv")
df.head(3)
```

### Bar Chart Ordering

Seaborn is a library that makes it easy to create bar charts with the `.countplot()` method.
#### 1. Nominal Data

```python
sns.countplot(df["victory_status"]);
```

```python
df["victory_status"].value_counts(ascending=True).index
```

```python
sns.countplot(df["victory_status"], order = df["victory_status"].value_counts(ascending=True).index);
```

#### 2. Ordinal Data

```python
df = pd.read_csv("school_data.csv")
df.head()
```

```python
df["Quality of Facilities"].value_counts()
```

```python
sns.countplot(df["Quality of Facilities"], order = ["NOT ENOUGH DATA","VERY WEAK", "WEAK", "NEUTRAL", "STRONG", "VERY STRONG"])
plt.xticks(rotation = 30)
plt.show()
```

### Pie Charts
Before we dive into pie chart use cases and how to plot them, we want to address the fact that pie charts are the most issue-ridden graphing visual.

The seaborn library does not currently have any pie chart libraries.

```python
water_usage = pd.read_csv("water_usage.csv")
water_usage
```

```python
plt.pie(water_usage["prop"], labels=  water_usage["water_source"])

plt.axis("equal")
plt.title("Distribution of House Water Usage")
plt.show()
```

```python
pie_data = pd.read_csv("pie_data.csv")
pie_data
```

```python
plt.subplot(1,2,1)
plt.pie(pie_data["values"], labels = pie_data["labels"])
plt.axis("equal")
plt.title("Tough to Compare")

plt.subplot(1,2,2)
sns.barplot(data = pie_data, x = "labels", y = "values")
plt.title("Easy to Compare")

plt.show()
```
